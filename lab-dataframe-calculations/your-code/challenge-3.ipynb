## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

```python
# your code here
#sub problem 1

orders_aggr=orders_df.groupby(["CustomerID"]).agg({"amount_spent":"sum"})
display(orders_aggr["amount_spent"])
```

```python
#sub problem 2
bins=pd.qcut(orders_aggr["amount_spent"],[0,.75,.95,1])
bins.value_counts()
```

```python
#sub problem 3
labels=["Preferred","VIP"]
bins=pd.qcut(orders_aggr["amount_spent"],[.75,.95,1],labels=labels)
bins
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
orders_df["bins"]=bins
orders_df
orders_VIP=orders_df[orders_df["bins"]=="VIP"]
orders_VIP_g=orders_VIP.groupby(["Country"]).agg({"bins":"count"})

orders_VIP_g.sort_values(by="bins",ascending=False)
#UK
```
